```python
def even_odd(data):
    if data%2==0:
        print(f"{data} is Even")
    else:
        print(f"{data} is odd")


idata=eval(input("Enter the number:"))
even_odd(idata)
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f" Positional argument: {val}")
    for key, value in kwargs.items():
        print(f"Key Argument : {key}:{value}")

print_details(1,2,3,4,5,"kunal",name="Rohit",marks=45)
```
