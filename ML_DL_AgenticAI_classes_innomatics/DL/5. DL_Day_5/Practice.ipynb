```python
import warnings
warnings.filterwarnings('ignore')

import cv2
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("python_img.png")  
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
```

```python
blur = cv2.GaussianBlur(img, (7,7), 0)
blur_rgb = cv2.cvtColor(blur, cv2.COLOR_BGR2RGB)

plt.imshow(blur_rgb)
plt.title("Blur Image")
plt.axis("off")
```

```python
edges = cv2.Canny(gray, 100, 200)

plt.imshow(edges, cmap='gray')
plt.title("Edge Detection")
plt.axis("off")
```

```python
img_copy = img.copy()

cv2.rectangle(img_copy, (50,50), (200,200), (0,255,0), 3)
cv2.putText(img_copy, "Demo", (50,40),
            cv2.FONT_HERSHEY_SIMPLEX, 1, (255,0,0), 2)

img_copy_rgb = cv2.cvtColor(img_copy, cv2.COLOR_BGR2RGB)

plt.imshow(img_copy_rgb)
plt.title("Rectangle + Text")
plt.axis("off")
```
