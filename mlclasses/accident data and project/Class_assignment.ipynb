2. View Students
3. Search Student
4. Update Marks
5. Delete Student
6. Exit

```python
class Student:
    def __init__(self, roll, name, marks):
        self.roll = roll
        self.name = name
        self.marks = marks


students = []


def add_student():
    roll = int(input("Enter roll: "))
    name = input("Enter name: ")
    marks = int(input("Enter marks: "))
    
    students.append(Student(roll, name, marks))
    print("Added!")


def view_students():
    for s in students:
        print(s.roll, s.name, s.marks)


def search_student():
    r = int(input("Enter roll: "))
    for s in students:
        if s.roll == r:
            print(s.roll, s.name, s.marks)
            return
    print("Not found")


def update_marks():
    r = int(input("Enter roll: "))
    for s in students:
        if s.roll == r:
            s.marks = int(input("New marks: "))
            print("Updated")
            return
    print("Not found")


def delete_student():
    r = int(input("Enter roll: "))
    for s in students:
        if s.roll == r:
            students.remove(s)
            print("Deleted")
            return
    print("Not found")


while True:
    print("\n1.Add 2.View 3.Search 4.Update 5.Delete 6.Exit")
    ch = int(input("Choice: "))

    if ch == 1:
        add_student()
    elif ch == 2:
        view_students()
    elif ch == 3:
        search_student()
    elif ch == 4:
        update_marks()
    elif ch == 5:
        delete_student()
    elif ch == 6:
        break
```
