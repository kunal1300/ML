# Reading large file

```python
def read_large_path(file):
    with open("large_file.txt","r") as file:
        for line in file:
            yield line
```

```python
file = read_large_path("large_file.txt")
```

```python
print(next(file))
```

```python
for line in file:
    print(line.strip())
```

```python
def repeat(n):
    def decorator (func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator
```

```python
@repeat(3)
def say_hello():
    print("hello",{"name":"Kunal"})
```

```python
say_hello()
```
